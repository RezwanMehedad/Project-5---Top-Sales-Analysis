- What is the overall sales trend?

```python
df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
df['month_year'] = df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
df_trend = df.groupby('month_year')['sales'].sum().reset_index()
```

```python
plt.figure(figsize=(15,6))

plt.plot(df_trend['month_year'], df_trend['sales'], color = 'red') 

plt.xticks(rotation = 'vertical', size=8)
plt.show()
```

Which are the top 10 products by sales?

```python
df.head()
```

```python
prod_sales = pd.DataFrame(df.groupby('product_name')['sales'].sum())
```

```python
prod_sales = prod_sales.sort_values('sales',ascending = False)
```

```python
# These are the most selling products.

prod_sales[:10]
```

Which are the most selling products?

```python
prod_quantity = pd.DataFrame(df.groupby('product_name')['quantity'].sum())
```

```python
prod_quantity = prod_quantity.sort_values('quantity', ascending = False)
```

```python
prod_quantity[:10]
```

```python
# The above table shows the top most selling products based on quantity sold
```

What is the most preferred ship mode?

```python
#Standard class is the most preferred ship mode

plt.figure(figsize = (10,8.5))
sns.countplot(data = df, x = 'ship_mode')
plt.show()
```
